```python
#What is re library used for?
#A regular expression or RegEx is a sequence of characters which enable you to find string or set of string using a specialized pattern. The module re provides support for Perl-like regular expressions in Python.

import re
df.loc[df['Name'].str.contains('^pi[a-z]*',flags=re.I,regex=True)]
```

```python
#reading modified file
df=pd.read_csv("modified.csv")
#it groub all the type 1 column value merged 
df.groupby(['Type 1']).mean().sort_values('Defense',ascending=False)
```

```python
#.loc will read the rows then it will check whether type 1 is equal to grass  type2 ==poison and hp
#value is grater the 70 then it will reset index drop and place and make a new file
newdf=df.loc[(df['Type 1']=='Grass')&(df['Type 2'] == 'poison') & (df['HP'] >70)]
newdf.reset_index(drop=True, inplace=True)
newdf
newdf.to_csv("filtered.csv")
```

```python
#Why is Distplot used?
#distplot() is used to visualize the parametric distribution of a dataset.
f,ax=plt.subplots(figsize=(10,8))
ax=sns.distplot(x=df['Attack'],bins=10)
plt.show()
```

```python
#relation between hp and attack
#Subplots mean a group of smaller axes (where each axis is a plot) that can exist together within a single figure

fig, axs=plt.subplots(2,2,figsize=(12,12))
ax1=plt.subplot2grid((8,8),(0,0),rowspan=3,colspan=3)
ax2=plt.subplot2grid((8,8),(4,0),rowspan=3,colspan=3)
ax3=plt.subplot2grid((8,8),(0,4),rowspan=3,colspan=3)
ax4=plt.subplot2grid((8,8),(4,4),rowspan=3,colspan=3)

fig.tight_layout()
ax1.set_title("Plot1: HP AND ATTACKS", fontsize=19)
ax2.set_title("Plot2: HP AND ATTACKS", fontsize=19)
ax3.set_title("Plot3: HP AND ATTACKS", fontsize=19)
ax4.set_title("Plot4: HP AND ATTACKS", fontsize=19)

#plot 1
sns.regplot(x='HP',y='Attack',data=df,ax=ax1)
#plot 2
sns.regplot(x='HP',y='Attack',marker='^',fit_reg=False,color="green",data=df,ax=ax2)
#plot 3
sns.regplot(x='HP',y='Attack',marker='^',fit_reg=False,color="green",data=df,ax=ax3)
#plot 4
sns.regplot(x='HP',y='Attack',marker='^',color="green",data=df,ax=ax4)
plt.show()
```
