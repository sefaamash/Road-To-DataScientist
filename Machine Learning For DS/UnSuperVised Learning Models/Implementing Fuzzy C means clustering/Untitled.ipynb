```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from fcmeans import FCM
```

```python
iris=pd.read_csv('iris.csv')
iris
```

```python
iris1=iris.drop(['Unnamed: 0'],axis=1)
iris1
```

```python
Specie=pd.get_dummies(iris1['Species'] ,drop_first=False)
Specie
```

```python
iris1=pd.concat([iris1,Specie],axis=1)
iris1
```

```python
featured_col=['Sepal.Length','Sepal.Width','Petal.Length','Petal.Width']
y_featured=['setosa','versicolor','virginica']
x=iris1[featured_col]
y=iris1[y_featured]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)

```

```python
cluster_list=[2,3,4,5,6,7]
model=list()
for n_clusters in cluster_list:
    fcm=FCM(n_clusters=n_clusters)
    fcm.fit(x_train)
    fcm.predict
    
```
