```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.preprocessing import scale
from sklearn.cluster import KMeans
```

```python
%matplotlib.inline
rcParams['figure.figsize']=7,8
```

```python
iris=datasets.load_iris()
iris.data
```

```python
#scaling
What does Sklearn scale do?
Standardize a dataset along any axis. Center to the mean and component wise scale to unit variance.
Why do we scale data?
Image result for what scale do in sklearn
Feature scaling is essential for machine learning algorithms that calculate distances between data. ... Since the range of values of raw data varies widely, in some machine learning algorithms, objective functions do not work correctly without normalization.
```

```python
x=scale(iris.data)
y=pd.DataFrame(iris.target)#It converts an array of taret column in to a pandas daatframe
variablesnames=iris.feature_names
x
```

```python
dataBunch
Dictionary-like object, with the following attributes.

data{ndarray, dataframe} of shape (150, 4)
The data matrix. If as_frame=True, data will be a pandas DataFrame.

target: {ndarray, Series} of shape (150,)
The classification target. If as_frame=True, target will be a pandas Series.

feature_names: list
The names of the dataset columns.

target_names: list
The names of target classes.
```

```python
#BUILDING AND RUNNING THE MODEL
#n_clusters=3 means how many centroid to chose each of the value will e compared to the centrod and the nearest to any centroid will be fall in that centrpoid category
clustering=KMeans(n_clusters=3,random_state=3)
clustering.fit(x)
```

```python
#PLOTTING THE MODEL
iris_df=pd.DataFrame(iris.data)
iris_df.columns=['Sepal Length','Sepal Width','Petal Length','Petal Width']
y.columns=['Targets']
```

```python
colortheme=np.array(['darkgray','lightsalmon','powderblue'])
#Actual Petal length and Width and target variable
plt.subplot(1,2,1)
plt.title('GROUND TRUTH CLASSIFICATION')
plt.scatter(x=iris_df['Petal Length'],y=iris_df['Petal Width'],c=colortheme[iris.target],s=50)
plt.legend()
#Predicted Petal length and Width and target variable
plt.subplot(1,2,2)
plt.title('K MEANS CLASSIFICATION')
plt.scatter(x=iris_df['Petal Length'],y=iris_df['Petal Width'],c=colortheme[clustering.labels_],s=50,label=['setosa','versicolor','virginica'])
plt.legend()
```

```python
relabel=np.choose(clustering.labels_,[2,0,1]).astype(np.int64)
plt.subplot(1,2,1)
plt.title('GROUND TRUTH CLASSIFICATION')
plt.scatter(x=iris_df['Petal Length'],y=iris_df['Petal Width'],c=colortheme[iris.target],s=50)
plt.legend()
#Predicted Petal length and Width and target variable
plt.subplot(1,2,2)
plt.title('K MEANS CLASSIFICATION')
plt.scatter(x=iris_df['Petal Length'],y=iris_df['Petal Width'],c=colortheme[clustering.labels_],s=50,label=['setosa','versicolor','virginica'])
plt.legend()
```
