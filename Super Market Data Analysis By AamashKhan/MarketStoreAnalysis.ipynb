```python
sales=pd.read_csv(r"C:\Users\Aamash\Desktop\Data Analysis Projects\Super Market Data Analysis By AamashKhan\supermarket_sales.csv")
sales.head()
```

```python
#dropping null areas
sales=sales.dropna(how='all')
sales.head(1000)
```

```python
#Changing to numeric
sales['Quantity']=pd.to_numeric(sales['Quantity'])
sales['Unit price']=pd.to_numeric(sales['Unit price'])
```

```python
#Making new Column
sales=sales[sales['Date'].str[0:1]!='or']
sales['Month']=sales['Date'].str[0:1]#from date we finding out month number and making a new column of it
sales.head(100)
```

```python
#Which Gender comes the most in the store
sns.set(style="darkgrid")
gendercount=sns.countplot(x="Gender",data=sales).set_title("Gender")
```

```python
#Checking which branch has the most ratings
sns.boxplot(x="Branch",y="Rating",data=sales).set_title("Branch Rating")
```

```python
#Sales Per Hour but first we have to fing the hour from date
type(sales['Date'])
sales['Date']=pd.to_datetime(sales['Date'])#from date taking out datetime
sales['Day']=(sales['Date']).dt.day#datetime to day
sales['month']=(sales['Date']).dt.month#to month
sales['year']=(sales['Date']).dt.year#to year
sales['Time']=pd.to_datetime(sales['Time'])# time taking out  datetime
sales['Hour']=(sales['Time']).dt.hour
#SALES PER HOUR
gendercount=sns.lineplot(x='Hour',y='Quantity',data=sales).set_title("Product Sales Per Hour")
```

```python
#First 3 months sales data of Three branches
gendercount=sns.relplot(x="Hour",y="Quantity",col="month",row="Branch",kind="line",hue="Gender",style="Gender",data=sales)
```

```python
#See the total with respect to ratings of the branch
ageSisspend=sns.lineplot(x="Total",y="Rating",data=sales)
```

```python
#Which product has the most sale
#order=sales['Product line'].value_counts().index[this is used for order from descending order the item which are sold the most are kept higher from sale lesser]
sns.countplot(y='Product line',data=sales,order=sales['Product line'].value_counts().index).set_title("Product line")
```

```python
#WE ARE FINDING THAT THE PRODUCTS WHICH ARE SOLD AT A HIGHER ARE BOUGHT BY MALES OR FEMALES THE MOST
sns.stripplot(y='Product line', x='Total',hue='Gender',data=sales)
```
