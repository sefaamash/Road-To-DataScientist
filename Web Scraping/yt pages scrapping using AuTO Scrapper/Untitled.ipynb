```python
import csv
from autoscraper import AutoScraper
```

```python
url=f"https://www.yellowpages.ca/search/si/1/hair+dresser/Karachi+Drive+Markham+ON"
wantedlist=['Jeya Beauty','7428 Markham Rd, Markham, ON L3S 4V6','289-554-1002']
scraper=AutoScraper()
result=scraper.build(url,wantedlist)
print(result)
```

```python
scraper.get_result_similar(url,grouped=True)
```

```python
scraper.set_rule_aliases({'rule_qdor':'Name','rule_ypp5':'Address','rule_bjm8':'Phone No'})
scraper.keep_rules(['rule_qdor','rule_ypp5','rule_bjm8'])
scraper.save('hairdresser')
```

```python
#pg2
results1=scraper.get_result_similar('https://www.yellowpages.ca/search/si/3/hair+dresser/Karachi+Drive+Markham+ON',group_by_alias=True)
```

```python
file=open("hairdresser_scrap.csv",'a')
writer=csv.writer(file)


for i,j,k in zip(results1['Name'],results1['Address'],results1['Phone No']):
    a=i
    b=j
    c=k
    print(a+" "+b+" "+c)

    writer.writerow ( [a,b,c])

```
